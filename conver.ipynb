```python
from transformers import MarianTokenizer, MarianMTModel
import pandas as pd
import torch
```

```python
import numpy as np

# Ensure the DataFrame has 5000 rows
df_trimmed = df.iloc[:5000]

# Convert the DataFrame to a numpy array
arr = np.array(df_trimmed)

# Split the array into 100 sub-arrays
arrs = np.array_split(arr, 1000)
```

```python
import torch
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast

# Load the model
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")

# If a GPU is available, load the model onto the GPU
if torch.cuda.is_available():
    model = model.to('cuda')

def e_h(article_hi, model=model):
    tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
    tokenizer.src_lang = "en_XX"
    encoded_hi = tokenizer(article_hi, return_tensors="pt")

    # If a GPU is available, load the tensors onto the GPU
    if torch.cuda.is_available():
        encoded_hi = encoded_hi.to('cuda')

    generated_tokens = model.generate(**encoded_hi, forced_bos_token_id=tokenizer.lang_code_to_id["hi_IN"])
    return tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)

hi = "It is very cold tonight."
print(e_h(hi, model))
```

```python
import threading

results = [None] * 25

def thread_function(thread_id):
    print('thread id', thread_id)
    df_first_10 =  pd.DataFrame(arrs[thread_id])
    df_first_10[3] = df_first_10[0].apply(e_h)
    results[thread_id] = df_first_10

threads = []
for i in range(0, 25):
    thread = threading.Thread(target=thread_function, args=(i,))
    threads.append(thread)
    thread.start()

for thread in threads:
    thread.join()

print("All threads have finished execution.")
result_df = pd.concat(results)
```
